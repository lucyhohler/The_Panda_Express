## Raw data
The purpose of this notebook is to take the raw data and convert it into two csv files. The raw data is located in the "Raw" directory. The converted data will be written to the "Data" directory. 

```python
# dependencies
import pandas as pd
import os
```

```python
# Creating a new csv that combines all the Raw NFL Data
output = "Data/NFL_Combined_2010_2015.csv"

df = pd.DataFrame()

nfl_dir = "./Raw/NFL_Combined/"
filenames = os.listdir(nfl_dir)  

for filename in filenames:
    df = df.append(pd.read_csv(nfl_dir + filename), ignore_index=True)
    
#Replacing names of positions for easier data analysis
#Outside Linebacker & Inside Linebacker converted to Linebacker allocation
df['Pos'] = df['Pos'].replace({'OLB': 'LB', 'ILB': 'LB', 'OG': 'OL', 'OT': 'OL', 'DE': 'DL', 'DT': 'DL', 'CB': 'DB', 'SS': 'DB', 'FS': 'DB'})    

df.to_csv(output, index=False)
df.head()
```

```python
# Creating a new csv that combines all the Raw NFL Pro Bowl Data
output = "Data/Pro_Bowl_2011_2018.csv"

df = pd.DataFrame()

nfl_dir = "./Raw/Pro_Bowl/"
filenames = os.listdir(nfl_dir)  

for filename in filenames:
    df = df.append(pd.read_csv(nfl_dir + filename), ignore_index=True)

#Cleaning ProBowl names; taking out the '+' and '%'
df.Player = [x.strip('+') for x in df.Player]
df.Player = [x.strip('%') for x in df.Player]
    
df.to_csv(output, index=False)
df.head(10)
```
